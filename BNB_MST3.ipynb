```python
import numpy as np
import os
import time
folder = "Problems"

time_limit = 300
```

```python
def get_mst(distances, Is_been_to):
    idx = [0]
    for i in range(1, len(Is_been_to)):
        if Is_been_to[i] == False:
            idx.append(i)
    n = len(idx)
    cost = [np.inf]*n
    cost[0] = 0
    cur = 0
    rest_locations = set(np.arange(1, n))
    for _ in range(n-1):
        min_cost = np.inf
        next = -1
        for i in rest_locations:
            cost[i] = min(cost[i], distances[idx[i]][idx[cur]])
            if cost[i] < min_cost:
                min_cost = cost[i]
                next = i
        rest_locations.remove(next)
        cur = next
    # print(cost)
    return sum(cost)
```

```python
result = [[] for _ in os.listdir(folder)]

for step, dirs in enumerate(sorted(os.listdir(folder), key = lambda x: int(x))):
    print("Test with number of {} begin!".format(dirs))
    cur_dir = os.path.join(folder, dirs)
    count = 0
    for files in os.listdir(cur_dir):
        start_time = time.time()
        path = os.path.join(cur_dir, files)
        with open(path, 'r', encoding='utf-8') as file:
            content = file.read().split()
            data = list(map(float, content))
            number = int(data[0])
            distances = np.zeros((number,number))
        for i in range(number):
            for j in range(number):
                distances[i][j] = data[i*number+j+1]
        distances_order = [sorted(np.arange(number), key = lambda x: distances[x][i]) for i in range(number)]
    
        Is_been_to = [False]*number
        cur_route = [0]
        ans = [np.inf]
        ans_route = [""]
        terminate = [False]
        def dfs(cur, step, cur_sum):

            if time.time() - start_time > time_limit:
                terminate[0] = True
            if step == number-1:
                
                if ans[0] > cur_sum + distances[cur][0]:
                    ans[0] = cur_sum + distances[cur][0]
                    cur_route.append(0)
                    ans_route[0] = "->".join(map(str, cur_route))
                    cur_route.pop()
                return
            
            
            Is_been_to[cur] = True
            hubrist = get_mst(distances, Is_been_to)
            for i in distances_order[cur]:
                if terminate[0]:
                    return 
                if Is_been_to[i] == False and cur_sum+distances[cur][i]+hubrist < ans[0]:             
                    cur_route.append(i)
                    dfs(i, step+1, cur_sum+distances[cur][i])
                    cur_route.pop()
            Is_been_to[cur] = False

        dfs(0,0,0)
        print("ans is: {}, route is: {}".format(np.round(ans[0], 2), ans_route[0]))
        result[step].append(ans[0])
    print()
    print()

    
```
