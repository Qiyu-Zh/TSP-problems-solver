```python
import networkx as nx
import os
import numpy as np
import time
time_limit = 10
```

```python
folder = "Problems"
method_idx = 1
tsp = nx.approximation.traveling_salesman_problem
can_mutiple_visit = False
stop = False
for dirs in sorted(os.listdir(folder), key = lambda x: int(x)):
    if stop:
        break
    print("Test with number of {} begin!".format(dirs))
    cur_dir = os.path.join(folder, dirs)
    for files in os.listdir(cur_dir):
        start_time = time.time()
        path = os.path.join(cur_dir, files)
        with open(path, 'r', encoding='utf-8') as file:
            content = file.read().split()
            data = list(map(float, content))
            number = int(data[0])
        G = nx.complete_graph(int(data[0]))
        
        for i in range(number):
            for j in range(number):
                if i != j:
                    G[i][j]['weight'] = data[i*number+j+1]
        # print(print(G[0][1]["weight"]))
        m2 = lambda G, wt: nx.approximation.simulated_annealing_tsp(G, "greedy", weight=wt, temp=500, max_iterations = 100)
        m3 = lambda G, wt: nx.approximation.threshold_accepting_tsp(G, "greedy", weight=wt, max_iterations = 100)
        methods = (None, m2, m3)

        route = tsp(G, method = methods[method_idx])
        if time.time() - start_time > time_limit:
            stop = True
            print("Time Limit Reach!!! No return for {}".format(number))
            break
        if can_mutiple_visit == False:
            visited_point = set()
            new_route = [0]
            for i in range(1, len(route)):
                if route[i] not in visited_point:
                    new_route.append(route[i])
                    visited_point.add(route[i])
            route = new_route
        ans = 0
         
        for i in range(number):
            ans += G[route[i]][route[i+1]]["weight"]
        print("ans is: {}, route is: {}".format(np.round(ans, 2), "->".join(map(str, route))))
    print()
    print()
            
```
